# **NumPy Broadcasting**

NumPy automatically adjusts array shapes so mathematical operations can work between arrays of different sizes.

````python
import numpy as np

arr = np.array([1,2,3])

print(arr + 5) # NumPy automatically converted:5 into [5,5,5] and then do addition [1,2,3] + [5,5,5] = [6,7,8] and This automatic expansion is called: Broadcasting
````

## **Why Broadcasting Is Amazing ?**

Without broadcasting, you'd need loops:

```
for i in range(len(arr)):
    arr[i] += 5
```

NumPy avoids loops internally.

This makes computations MUCH faster.

## **Broadcasting Rules**

Two arrays can work together if:

**Rule 1** : Dimensions are equal

OR

**Rule 2** : One of them is 1

````python
#Example 2 — Same Shape

a = np.array([1,2,3])
b = np.array([10,20,30])

print(a + b) # Here both array shape is (3,) and (3,) that,s why works perfectly.
````

````python
#Example 3 — Different Shapes

a = np.array([[1],
              [2],
              [3]])

b = np.array([10,20,30])

print(a + b) # Here Shape of a:(3,1)  &  b:(3,) . So NumPy automatically expands arrays.
````

## **Broadcasting With 2D Arrays**

````python
a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([10,20,30])

print(a + b) # Here Numpy repeated [10 20 30] for each row . Internally:[[10 20 30] ,[10 20 30]] Then addition happens.
````

## **When Broadcasting Failure ❌**

````python
a = np.array([1,2,3])
b = np.array([1,2])

print(a + b)  # Because shape are different one is (3,) and another one is (2,) . So broadcasting fails.
````

## **Broadcasting + reshape()**

VERY IMPORTANT COMBINATION.

````python
a = np.array([1,2,3]).reshape(3,1)

b = np.array([10,20,30])

print(a + b) # Dimensions Compare From RIGHT SIDE . So Numpy reads : (3,1) , (1,3) . Then check (3,1)->ok , (1,3)->ok.
````
