# **NumPy Statistical Functions**



## **⭐ 1. mean()**

Mean = Average

Formula

Mean=
∑x
/ n

````python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

mean = np.mean(arr)

print("Mean of those number is : ",mean)  #(10 + 20 + 30 + 40 + 50) / 5 = 30
````

## **⭐ 2. median()**

Median = Middle value after sorting.

````python
median= np.median(arr)

print("Median of those number is : ",median)
````

## **Example with Outlier**

## **Important Understanding**

Mean gets affected by outliers.

Median is more robust.

*This is VERY important in Data Science.*

````python
data = np.array([10,20,30,40,500])

print("Using Mean : ",np.mean(data))

print("Using Median : ",np.median(data))
````

## **⭐ 3. sum()**

Adds all values.

````python
sum = np.sum(arr)

print("Adding all values : ", sum)
````

## **⭐ 4. min() and max()**

Find smallest and largest values.

````python
arr = np.array([10, 20, 30, 40, 500])

min= np.min(arr)

print("Minimum value of array is : ",min)

max= np.max(arr)

print("Maximun value of array is : ",max)
````

## **⭐ 5. argmin() and argmax()**

Returns INDEX position.

````python
arr = np.array([10, 20, 30, 400, 5])

max_index = np.argmax(arr)

print("Index of Maximum value : ",max_index)

min_index = np.argmin(arr)

print("Index of Minimum value : ",min_index)
````

## **⭐ 6. std()**

Standard Deviation

Measures:
> “How spread out data is.”

Small std:
- data close together

Large std:
- data widely spread

````python
arr = np.array([10,20,30,40,50])

std = np.std(arr)

print("Standard Deviation of this array is : ",std)
````

## **⭐ 7. var() — Variance**

**Variance** = *square of standard deviation.*

**Relation** : *Variance = (Standard Deviation)^2*

````python
arr = np.array([10,20,30,40,50])

var = np.var(arr)

print("Variance of this array is : ",var)
````

## **⭐ 8. percentile()**

Very important in Data Science.

Percentile means:

> “What percentage of data lies below a value.”

Important Percentiles



```
| Percentile | Meaning |
| ---------- | ------- |
| 25th       | Q1      |
| 50th       | Median  |
| 75th       | Q3      |
```



````python
arr = np.array([10,20,30,40,50])

perc=np.percentile(arr,25)

print("25th percentile of this array is : " , perc)

perc=np.percentile(arr,50)

print("50th percentile of this array is : " , perc) #50th percentile = median.

perc=np.percentile(arr,75)

print("75th percentile of this array is : " , perc)

perc=np.percentile(arr,100)

print("100th percentile of this array is : " , perc) #50th percentile = last value.
````

## **⭐ 9. unique()**

## **⭐ 10. sort()**

# **Statistical Functions on 2D Arrays**

````python
data = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,90]])

print(np.mean(data))  #Mean of Entire Array

print(np.mean(data, axis=0))   #Column-wise Mean

print(np.mean(data, axis=1))   #Row-wise Mean
````

**Understand axis=0**

*Means: Go DOWN columns*

Calculation:

First column: `(10 + 40 + 70)/3 = 40`

Second: `(20 + 50 + 80)/3 = 50`

Third: `(30 + 60 + 90)/3 = 60`

---

**Understand axis=1**

*Means: Go ACROSS rows*

Calculation:

First row: `(10 + 20 + 30)/3 = 20`

Second: `(40 + 50 + 60)/3 = 50`

Third: `(70 + 80 + 90)/3 = 80`
