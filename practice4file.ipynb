# **Practice Exercises**

**Exercise 1**

`arr = np.array([5,10,15,20,25])`

Find:

- mean
- median
- std
- max

````python
import numpy as np
arr = np.array([5,10,15,20,25])

print("Mean of this array is : ",np.mean(arr))
print("Median of this array is : ",np.median(arr))
print("Standard Diviation of this array is : ",np.std(arr))
print("maximun value of this array is : ",np.max(arr))
````

**Exercise 2**

```
data = np.array([[10,20],
                 [30,40],
                 [50,60]])
```

Find:

- column mean
- row mean

````python
data = np.array([[10,20],
                 [30,40],
                 [50,60]])

col_mean = np.mean(data,axis=0)
row_mean = np.mean(data,axis=1)

print("Column Mean : ",col_mean)
print("Row Mean : ",row_mean)
````

**Exercise 3**

Find:

- 25th percentile
- 75th percentile

````python
data = np.array([[10,20],
                 [30,40],
                 [50,60]])

print("25th percentile : ",np.percentile(data,25))
print("75th percentile : ",np.percentile(data,75))
````

## **Mini Challenge**

Create:

`np.random.randint(1,100,(5,4))`

Then calculate:

- Overall mean
- Column mean
- Row mean
- Maximum value
- Standard deviation
- 75th percentile

````python
arr = np.random.randint(1,100,(5,4))
print(arr)
print("Overall mean : ",np.mean(arr))
print("Column mean : ",np.mean(arr,axis=0))
print("Row mean : ",np.mean(arr,axis=1))
print("Maximum value : ",np.max(arr))
print("Standard deviation : ",np.std(arr))
print("75th percentile : ",np.percentile(arr,75))
````
