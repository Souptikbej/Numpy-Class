# **NumPy Random Module**

## **⭐ 1. rand()**

Generates random decimal numbers between:

```python
import numpy as np

print(np.random.rand())  #Every run gives different values within 0 and 1.
```

Multiple Random Values

```python
print(np.random.rand(5))
```

2D Random Array

```python
print(np.random.rand(3,3))
```

## **⭐ 2. randint()**

Generates random INTEGER values.

```python
print("Single Integer : ",np.random.randint(1,10))

print("Multiple Integers : ",np.random.randint(1,10,5))

print("2D Integer Array  : \n",np.random.randint(1,100,(3,4)))
```

## **⭐ 3. seed()**

Controls randomness.

**Problem Without seed()**

- Every run gives different output.

- Sometimes we want SAME results.

```python
np.random.seed(42)

print(np.random.randint(1,100,5))
```

## **⭐ 4. choice()**

Randomly selects values.

```python
arr = [10,20,30,40]

print(np.random.choice(arr))

print(np.random.choice(arr, 3))  #Multiple Choices
```

## **⭐ 5. shuffle()**

Shuffles ORIGINAL array.

```python
arr = np.array([1,2,3,4,5])

np.random.shuffle(arr) #shuffle() changes original array.

print(arr)
```

## **⭐ 6. permutation()**

Similar to shuffle().

BUT:

- does NOT modify original array

```python
arr = np.array([1,2,3,4,5])

new_arr = np.random.permutation(arr)

print(new_arr)
print(arr)
```

## **⭐ 7. uniform()**

Random decimal values between range.

```python
print(np.random.uniform(1,10,5))
```

## **⭐ 7. normal()**

Generates values from Normal Distribution.

Very important in AI/ML.

**Simple Meaning**

*Creates data around a mean.*

*Most real-world data follows normal distribution.*

Examples:

- Heights
- Marks
- Salaries
- Sensor data

```python
data = np.random.normal(50, 10, 5) # (mean, standard_deviation, total_values)

print(data) #50 = center , 10 = spread , 5 = number of values
```
