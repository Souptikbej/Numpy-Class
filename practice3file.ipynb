## **Mini Challenge**

Create:



```
np.random.randint(1,100,20)
```



Then:

Find even numbers

Find odd numbers

Find values > 50

Remove values < 20

Replace values > 80 with 100 using np.where()

````python
arr = np.random.randint(1,100,20)
print(arr)

print("\nEven numbers:")
print(arr[arr % 2 == 0])

print("\nOdd numbers:")
print(arr[arr % 2 != 0])

print("\nValues > 50:")
print(arr[arr > 50])

print("\nValues < 20:")
print(arr[arr < 20])

arr = np.where(arr > 80, 100, arr)
print("\nValues > 80 replaced with 100:")
print(arr)
````
