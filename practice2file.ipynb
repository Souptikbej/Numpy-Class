Understand HOW broadcasting happens.

````python
a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([10,20,30])

print(a + b) # Here a shape is (2,3) & b shape is (3,) . So b internly convert into [[10,20,30],[10,20,30]] , now shape is (2,3) . In that way Broadcasting happend easily.
````
